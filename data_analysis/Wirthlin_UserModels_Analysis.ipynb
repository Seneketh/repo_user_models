```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import csv
import time
import datetime
from random import randint
import glob
import os
from scipy import stats
```

Data Processing for Baseline Condition Subject C

```python
data = pd.read_csv("MERGED.csv").astype('object')

allpupildata = []
levelgravity = []
levelthreshcount = []
leveldeath = []

for current_line in range(len(data.level_pupdilations)):
    if len(data.loc[current_line, 'level_pupdilations']) > 5: #getting rid of the strange "missing data"
        tempdata = data.loc[current_line, 'level_pupdilations']
        tempdata  = tempdata[1:-1] #only with python lists (to remove the brackets)
        tempdata = tempdata.split(',')
        data.loc[current_line, 'level_pupdilations'] = map(float, tempdata)

        if data.baselining[current_line] == True and data.control[current_line] == False and data.ID[current_line] == 'c':
            allpupildata.append(data.loc[current_line, 'level_pupdilations'])
            levelgravity.append([data.loc[current_line, 'gravity']]*60)
            levelthreshcount.append([data.loc[current_line, 'threscount']]*60)
            leveldeath.append([data.loc[current_line, 'deaths']]*60)

flatted_allpupildata = []
flatted_allpupildata.append([item for sublist in allpupildata for item in sublist if item > 9000])
flatted_allpupildata = flatted_allpupildata[0]
flatted_allpupildata = stats.zscore(flatted_allpupildata)

flatted_levelgravity = []
flatted_levelgravity.append([item for sublist in levelgravity for item in sublist])
flatted_levelgravity = flatted_levelgravity[0]
flatted_levelgravity = stats.zscore(flatted_levelgravity)
```

Baseline Plot Subject C:

```python
with plt.style.context('ggplot'): 
    fig, ax = plt.subplots()
    ax.plot(flatted_levelgravity[0:(20*60)], label='Gravity')
    ax.plot(flatted_allpupildata[0:(20*60)], label='Dilation', alpha=0.6)

    # Now add the legend with some customizations.
    legend = ax.legend(loc='upper right')

    plt.xlabel('Measurements (30 per Second)', fontsize=17)
    plt.ylabel('Z-Transformed Values' , fontsize=17)
    plt.title('Baselining Phase Subject C' , fontsize=25)

    plt.show()
```

Plotting Mean Task performance among condition:

```python
data = data[data.ID != 'f']
ax = sns.barplot(data=data, x="ID", y='deaths', hue='control')
ax.set(xlabel='Subjects', ylabel='Mean Game Losses')
ax = sns.plt.title('Task Performance per Subject')
sns.plt.show()
```
